```python
%sql
SELECT 
    request_params.full_name_arg AS tabela,
    COUNT(*) AS acessos
FROM system.access.audit
WHERE event_date >= current_date() - 30
  AND request_params.full_name_arg LIKE 'fin_prod%'
GROUP BY 1
ORDER BY acessos DESC
```

```python
%sql
SELECT usage_date, sku_name, SUM(usage_quantity) AS total
FROM system.billing.usage
WHERE usage_date >= current_date() - 30
GROUP BY 1, 2
ORDER BY 1 DESC
```

```python
%sql
SELECT 
    CASE 
        WHEN request_params.full_name_arg LIKE '%bronze%' THEN '1_bronze'
        WHEN request_params.full_name_arg LIKE '%silver%' THEN '2_silver'
        WHEN request_params.full_name_arg LIKE '%gold%'   THEN '3_gold'
        ELSE '9_outros'
    END AS camada,
    COUNT(*) AS total_acessos
FROM system.access.audit
WHERE event_date >= current_date() - 30
  AND request_params.full_name_arg LIKE 'fin_prod%'
GROUP BY 1
ORDER BY 1
```
